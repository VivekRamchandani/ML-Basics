```python
def cosine(x, y):
    xy = np.dot(x, y)
    len_x = np.sqrt((x**2).sum())
    len_y = np.sqrt((y**2).sum())

    return xy / (len_x * len_y)
```

```python
cosine(df["Angelica"], df["Veronica"])
```

```python
cosine_matrix = pd.DataFrame()
for x in df.columns:
    for y in df.columns:
        cosine_matrix.loc[x, y] = cosine(df[x], df[y])

cosine_matrix
```
