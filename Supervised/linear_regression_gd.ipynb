# Linear Regression using Gradient Descent

```python
import numpy as np
from sklearn.datasets import make_regression
import seaborn as sns
```

```python
class LinearR:
    def __init__(self):
        self.intercept: np.float64 = 0
        self.slope: np.float64 = 0
        self.cost_history = []
    
    def fit(self, x, y, l_rate=0.1, epoch=100):
        """
        Args:
            x (array-like): Independent values
            y (array-like): dependent values
            l_rate (float): Learning rate
            epoch (int): num of times to run loop
            
        """
        
        X = np.array(x)
        Y = np.array(y)
        m = len(X)
        # print(X.shape, Y.shape)
        X_b = np.c_[np.ones((m, 1)), X]
        theta = np.zeros(2)

        for i in range(0, epoch):    
            pred = np.dot(X_b, theta)
            theta = theta - (l_rate/m) * 2 * np.dot(X_b.T, (pred - Y))

        self.intercept, self.slope = theta[0], theta[1]

        return self
```

```python
x, y = make_regression(n_samples=50, n_features=1, noise=20, random_state=42)
model = LinearR().fit(x, y)
```

```python
print(model.intercept, model.slope)
```

```python
sns.set_theme()
sns.regplot(x=x.flatten(), y=y)
```
